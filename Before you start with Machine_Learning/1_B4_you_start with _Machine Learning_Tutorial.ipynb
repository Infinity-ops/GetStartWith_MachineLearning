# 2. Introduction to Numpy

Numpy is a library in python that specializes in dealing with multidimensional Arrays. The cool features of Numpy are

- **Automatic Checking:** Numpy ndArrays automatically check  the consistancy of data. For instance, it is not possible to have 1st row with 2 elements and 2nd row with 3 elements.
- **Contiguous Storage:** Unlike Python Lists, Numpy stores the data in contiguous Memory Locations, leading to lesser space.
- **Faster Vector Arithmetics:**  Because of Contiguous Storage, the operations are performed faster as compared to default Phython Execution for Lists.

# 2.2 Create your first plot in Numpy

```python
# Importing matplotlib Library
import matplotlib.pyplot as plt
```

```python
a = np.arange(10)

## Note:: np.arange( start, end, step)
# single argument - 0,end
# Two arguments - start, end
# Three arguments - start, end, step
```

```python
plt.plot(a)  #straight line
```

```python
plt.plot(a*a) #parabola
```

```python
# Equation of line
x = np.arange(-4,4,1)

# y = mx + c
y = 2*x + 5
```

```python
plt.plot(x,y)
plt.grid(True)
```

```python
plt.scatter(x,y)
plt.grid(True)
```

```python
x = np.arange(100)
y = np.sin(x)      # try out for cos, exp, log(log won't work as x has 0 value :) )
```

```python
plt.plot(x,y)
```

# 3. Random numbers

```python
np.random.random()   # Execute two or three times to see the difference (it generate from 0-1)
```

```python
 np.random.randint(0,3)  # This one generate random number between 0-3 (not 3) only integers
```

```python
np.random.randint(0,5,(10,))   # To generate array of random numbers
```

```python
# 1.  Lets plot this one
a = np.random.randint(0,5,(10,))
plt.plot(a)
```

```python
# Above plot doesn't give any meaning
# so to check the random number Let's try Histogram

## 2.  Histogram
a = np.random.randint(0,5,(100,))
plt.hist(a)
```

```python
plt.hist(a, bins = 5)
## NOTE:
# where you can see every numbers has equal distribution
# That's why its called uniform Random Number Generation
```
